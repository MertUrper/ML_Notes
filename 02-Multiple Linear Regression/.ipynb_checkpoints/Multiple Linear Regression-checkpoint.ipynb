# Train - Test Split

DatamÄ±z 1000'e kadar ise % 20-30 test, % 70-80 train

DatamÄ±z 1000 - 100.000 aralÄ±ÄŸÄ±nda ise % 10-20 test, % 80-90 train

DatamÄ±z 100.000 - 1.000.000 aralÄ±ÄŸÄ±nda ise % 5 test, % 95 train

DatamÄ±z + 1.000.000 ise % 1 test, % 99 train

en dÃ¼ÅŸÃ¼k oran ise %30 test, %70 train olabilir. Train datasÄ±nÄ±n fazla olmasÄ± Ã¶nemli

# Adjusted R2 Score (dÃ¼zeltilmiÅŸ R2 skor)

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def adj_r2(y_test, y_pred, df):
    r2 = r2_score(y_test, y_pred)
    n = df.shape[0]                # gÃ¶zlem 
    p = df.shape[1] -1             # feature
    adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p -1)
    return adj_r2
adj_r2(y_test, y_pred, df)
```

```python
def adj_r2(y_test, y_pred, df):
    r2 = r2_score(y_test, y_pred)
    n = df.shape[0]
    p = 100
    adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p -1)
    return adj_r2
adj_r2(y_test, y_pred, df)
# sÃ¼tun sayÄ±sÄ± artarken satÄ±r sayÄ±sÄ± da artmalÄ± yoksa tahmin gÃ¼cÃ¼mÃ¼z  azalabilir
```

# What is the concept of punishment for RMSE error metric?

```python
variables = {"Actual": [1000, 2000, 3000, 4000, 5000], "pred" : [1100, 2200, 3200, 4200, 5300]}
variables2 = {"Actual": [1000, 2000, 3000, 4000, 5000], "pred" : [1100, 2200, 3200, 4200, 5800]}
```

```python
df_2 = pd.DataFrame(variables)
df_2
```

```python
df_3 = pd.DataFrame(variables2)
df_3
```

```python
df_2["residual"] = abs(df_2.Actual - df_2.pred)

df_2
```

```python
df_3["residual"] = abs(df_3.Actual - df_3.pred)

df_3
```

```python
df_2.residual.sum() / 5    
```

```python
((df_2.residual ** 2).sum() / 5) ** 0.5
```

```python
df_3.residual.sum() / 5
```

```python
((df_3.residual ** 2).sum() / 5) ** 0.5
```

# Evaluation Model

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    print("Model testing performance: ")
    print("--------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
```

```python
eval_metric(y_test, y_pred) #test datasÄ± sonuÃ§larÄ±
```

```python
y_train_pred = model.predict(X_train)
```

```python
eval_metric(y_train, y_train_pred) #train datasÄ± sonuÃ§larÄ±
```
