```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#%matplotlib inline
#%matplotlib notebook
plt.rcParams["figure.figsize"] = (10, 6)
import warnings

warnings.filterwarnings("ignore")
warnings.warn("this will not show")
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df = pd.read_csv("mouse_viral_study.csv")
```

The data set simulates a medical study in which mice infected with the virus were treated with two different doses of drugs to observe whether the mice were infected with the virus after two weeks.

```python
df.head()
```

```python
df.info()
```

# Separating Hyperplane Manually

```python
sns.scatterplot(x = 'Med_1_mL', y = 'Med_2_mL', hue = 'Virus Present', data = df);
x = np.linspace(0, 10, 100)
m = -1
b = 11
y = m * x + b
plt.plot(x, y, 'black')
plt.show()
```

# SVM - Support Vector Machine

```python
from sklearn.svm import SVC
```

```python
help(SVC)
```

NOTE: For this example, we will explore the algorithm, so we'll skip any scaling or even a train\test split for now

```python
X = df.drop('Virus Present', axis = 1)
y = df['Virus Present']
```

```python
model = SVC(kernel = 'linear', C = 1)
model.fit(X, y)
```

```python
# https://scikit-learn.org/stable/auto_examples/svm/plot_separating_hyperplane.html

from svm_margin_plot import plot_svm_boundary  # This is imported from the supplemental .py file
# kendi yaptığımız code'u import ettik
```

```python
plot_svm_boundary(model, X, y)
```

# Hyper Parameters

## C
Regularization parameter. The strength of the regularization is inversely proportional to C. Must be strictly positive. The penalty is a squared l2 penalty.

```python
model = SVC(kernel = 'linear', C = 0.01)
model.fit(X, y)
plot_svm_boundary(model, X, y)
```

## Gamma

klasların etrafına set çiziyor

```python
model = SVC(kernel='rbf', C = 1.5, gamma = "scale")
model.fit(X, y)

plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel = 'sigmoid')
model.fit(X, y)

plot_svm_boundary(model, X, y)
```

## Degree (poly kernels only)

```python
model = SVC(kernel = 'poly', C = 1, degree = 5)
model.fit(X, y)

plot_svm_boundary(model, X, y)
```

# Grid Search

```python
from sklearn.model_selection import GridSearchCV
svm = SVC()
param_grid = {
    'C': [0.0001, 0.01, 0.1],
    'kernel': ['linear', 'rbf', 'sigmoid', 'poly'],
    'gamma': ["scale", "auto"],
    'degree': [1, 2]
}
grid = GridSearchCV(svm, param_grid)
```

```python
grid.fit(X, y)
```

```python
grid.best_score_
```

```python
grid.best_params_
```
