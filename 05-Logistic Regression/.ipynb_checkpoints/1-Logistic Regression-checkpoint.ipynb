# Train | Test Split and Scaling

```python
X = df.drop(columns = "test_result")
y = df.test_result
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Modelling

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
log_model.fit(X_train_scaled, y_train)
```

```python
log_model.coef_
```

```python
log_model.intercept_
```

```python
y_pred = log_model.predict(X_test_scaled)
y_pred
```

```python
y_pred_proba = log_model.predict_proba(X_test_scaled)
y_pred_proba

## Classification modellerinede predict_proba diye birşey var.
 # birinci sütun 0 olma olasılığı ikinci sütun 1 olma olasılığı.
 # Bu işlemi kendi çapında olasılık alıp yapıyor.
```

# X_test + y_yest + y_pred + y_pred_proba

```python
test_data = pd.concat([X_test, y_test], axis=1)
test_data
```

```python
## Orjinal datama probaları ilave ediyorız.
 # Burada 0.5 ve üzerindeki olasılıkları 1 classına,
 #        0.5 ve altındakileri de 0 classına atar.

test_data["pred_proba"] = y_pred_proba[:,1]
test_data
```

```python
## Buradan bakınca da aynı şey görünüyor.
 # Hedef classının hep 1 olması iyidir.
test_data["pred"] = y_pred
test_data
```

```python
test_data.sample(10)
```

```python
# Kaçta kaç bulmuş:

test_data.groupby("test_result")[["pred"]].value_counts(normalize = True)
```

# Final Model and Model Deployment

```python
scaler = StandardScaler().fit(X)

## Modellerimizi kaydederken sadece fit edilmiş kısmı ile kaydediyoruz.
 # Fit transform yaptıktan sonra kaydedersek, sonradan tahmin ettireceğimiz datadan hata alırız.
```

```python
## Modellerimizi kaydederken sadece fit edilmiş kısmı ile kaydediyoruz.
 # Fit transform yaptıktan sonra kaydedersek, sonradan tahmin ettireceğimiz datadan hata alırız.

import pickle
pickle.dump(scaler, open("scaler_hearing", 'wb'))

## Hangi isimle kaydetmek istiyorsak ilk önce onu yazıyoruz.
## Sadece w yazsakta kaydeder, biz localimizde az yer kaplaması için wb yazdık.
```

```python
X_scaled = scaler.transform(X) ## Datamın hepsine transform uyguluyorum.
```

```python
final_model = LogisticRegression().fit(X_scaled, y)
```

```python
## Şimdi de final modelimizi kaydediyoruz. 

pickle.dump(final_model, open("final_model_hearing", 'wb'))
```

```python
my_dict = {"age": [20, 30, 40, 50, 60],
           "physical_score": [50, 30, 20, 10, 5]}

## Test etmek için rastgele değerler aldık.

sample = pd.DataFrame(my_dict)
sample
```

```python
scaler_hearing = pickle.load(open("scaler_hearing", "rb"))

## Kaydettiğimiz datayı çağırıyoruz. Bu sefer rb yazdık.
```

```python
## Yukarıdaki modelimiz scale edilmiş model olduğu için, bizde sample değerlerimizi scale ediyoruz.

sample_scaled = scaler_hearing.transform(sample)
sample_scaled
```

```python
final_model = pickle.load(open("final_model_hearing", "rb"))

## Kaydettiğimiz modelimizi çağırıyoruz.
```

```python
predictions = final_model.predict(sample_scaled)
predictions_proba = final_model.predict_proba(sample_scaled)
```

```python
sample["pred"] = predictions
sample["pred_proba"] = predictions_proba[:,1]
sample
```
