# Principal Component Analysis (PCA) - Unsupervised Learning

Principal Component Analysis (PCA), unsupervised bir öğrenme tekniğidir ve birçok pratik uygulamaya sahiptir. Bu teknik, veri setlerindeki değişkenler arasındaki ilişkiyi anlamak için kullanılır.

## Avantajları:

- **Korelasyonun Azaltılması:** PCA, çoklu doğrusallık (multicollinearity) sorununu ele alabilir.
- **Veri Sıkıştırma:** Veri setinin boyutunu azaltarak hesaplama maliyetini düşürmeye yardımcı olur.
- **Görselleştirme:** Yüksek boyutlu veri setlerini 2D veya 3D'de görselleştirmek için kullanılır.
- **Overfitting'e Dayanıklılık:** PCA, modelin veriye aşırı uyum (overfitting) yapmasına karşı dayanıklıdır.

## Sınırlamaları:

- **Lineerlik:** PCA lineer bir tekniktir. Bu nedenle doğrusal olmayan yapıları yakalamakta başarısız olabilir.
- **Maksimum Varyans Yaklaşımı:** Maksimum varyansı koruma yaklaşımı, her zaman en iyi özellikleri seçmek için en iyi yöntem olmayabilir.
- **Öz Değerlere Dayalı Bileşen Seçimi:** PCA, öz değerlere dayanarak bileşen seçimi yaptığı için bazen küçük öz değerli bileşenlerdeki önemli bilgileri göz ardı edebilir.

## PCA ve Ölçeklendirme

### Ölçeklendirmenin Önemi

PCA'nın ana hedefi verideki varyansı yakalamaktır ve bu varyansın anahtarı, özellikler arası ilişkileri yansıtan kovaryans matrisidir. Özelliklerin ölçeklendirilmesi, kovaryans matrisinin ve sonuçta elde edilen ana bileşenlerin nasıl şekilleneceğini doğrudan etkiler. Bu yüzden PCA'de özellikleri ölçeklendirirken dikkatli olmalıyız.

### Standartlaştırma (Standard Scaling) ve Min-Max Ölçeklendirme

Veriyi ölçeklendirmek için en yaygın kullanılan iki yöntem Standartlaştırma ve Min-Max ölçeklendirmedir:

- **Standartlaştırma:** Her özelliği, ortalamasının 0 ve standart sapmasının 1 olacak şekilde ölçekler. Bu, özellik değerlerini genellikle -3 ile +3 arasında bir aralığa çeker (ancak bu aralıkla sınırlı değildir) ve özellikler arasındaki varyansı korur.
  
- **Min-Max Ölçeklendirme:** Her özelliği genellikle [0,1] aralığına ölçekler. Bu, verinin minimum ve maksimum değerleri temel alınarak yapılan bir dönüşümdür. Bu yöntem, özellikler arasındaki mutlak değerleri korumak istediğinizde yararlı olabilir, ancak varyansı korumaz.

### PCA İçin Hangi Ölçeklendirme Yöntemini Kullanmalıyız?

PCA'nın odaklandığı ana şey varyanstır. Standartlaştırma, özellikler arasındaki orijinal varyansı korurken, Min-Max ölçeklendirme varyansı korumaz. Bu nedenle, PCA uygularken standartlaştırma genellikle tercih edilir. Min-Max ölçeklendirme, PCA'nın veri üzerindeki varyansı doğru bir şekilde yakalamasını engelleyebilir.

# PCA'nın `fit_transform` Metodu

`principal_components = pca.fit_transform(scaled_X)` koduyla PCA algoritmasının iki ana adımını gerçekleştiriyoruz: `fit` ve `transform`.

## **1. Fit Adımı:**

Bu adımda PCA, veriye "uyar" (fit) ve ana bileşenleri bulur. Bu süreçte şunlar gerçekleşir:

- **Merkezleştirme:** Özelliklerin ortalaması hesaplanır ve her özellikten bu ortalamalar çıkarılır.
- **Kovaryans Matrisinin Hesaplanması:** Merkezleştirilmiş veri üzerinden kovaryans matrisi hesaplanır.
- **Özdeğerler ve Özvektörlerin Hesaplanması:** Kovaryans matrisinin özdeğerleri ve özvektörleri hesaplanır. Özvektörler, ana bileşenlerin yönlerini belirtir; özdeğerler ise bu bileşenlerin verideki varyansını belirtir.
- **Ana Bileşenlerin Seçilmesi:** Eğer tüm ana bileşenler istenmiyorsa (örn. PCA'ya bileşen sayısı olarak 2 verildiyse) en büyük özdeğerlere sahip özvektörler seçilir.

## **2. Transform Adımı:**

Bu adımda PCA, orijinal veriyi yeni ana bileşenler üzerinden bir temsile dönüştürür:

- **Verinin Yeniden Boyutlandırılması:** `fit` adımında bulunan özvektörler (ana bileşenlerin yönleri) kullanılarak veri yeni bir uzaya dönüştürülür. Bu dönüşüm, orijinal veri matrisi ile özvektörlerin çarpılmasıyla elde edilir.

`fit_transform` metodu bu iki adımı sırayla gerçekleştirir ve sonuçta veriyi ana bileşenler üzerinden yeni bir temsile dönüştürmüş olur.

> **Not:** Eğer daha sonra aynı PCA modeliyle yeni bir veri kümesini dönüştürmek isterseniz, bu yeni veri üzerinde sadece `transform` metodunu kullanabilirsiniz. Çünkü `fit` adımı sadece bir kez, ilk veri kümesi üzerinde yapılır.

### PCA'da `pca.explained_variance_ratio_` ve `pca.explained_variance_ratio_.cumsum()`

`pca.explained_variance_ratio_` kodu, her ana bileşenin toplam varyans içerisinde ne kadarlık bir kısmı açıkladığını belirtir.

- **explained_variance_ratio_**: Her bileşenin toplam varyans içerisindeki oransal katkısını gösterir. Değerler en çok varyansı açıklayan bileşenden en aza doğru sıralanır.

Örneğin: [0.7, 0.2, 0.1]

çıktısında ilk bileşen varyansın %70'ini, ikincisi %20'sini ve üçüncüsü %10'unu açıklar.

`pca.explained_variance_ratio_.cumsum()` ise bileşenlerin kümülatif (birikimli) oransal katkısını gösterir. Yani ilk bileşenin katkısı, ilk iki bileşenin toplam katkısı, ilk üç bileşenin toplam katkısı şeklinde devam eder.

Bu bilgiler, veri setinin boyutunu azaltırken hangi bileşenlerin korunması gerektiğine karar vermek için önemlidir.

## K-Means Algorithm

```python
from sklearn.neighbors import BallTree
import numpy as np
import pandas as pd
def hopkins(data_frame, sampling_size):
    """Assess the clusterability of a dataset. A score between 0 and 1, a score around 0.5 express
    no clusterability and a score tending to 0 express a high cluster tendency.
    Parameters
    ----------
    data_frame : numpy array
        The input dataset
    sampling_size : int
        The sampling size which is used to evaluate the number of DataFrame.
    Returns
    ---------------------
    score : float
        The hopkins score of the dataset (between 0 and 1)
    Examples
    --------
    >>> from sklearn import datasets
    >>> from pyclustertend import hopkins
    >>> X = datasets.load_iris().data
    >>> hopkins(X,150)
    0.16
    """
    if type(data_frame) == np.ndarray:
        data_frame = pd.DataFrame(data_frame)
    # Sample n observations from D : P
    if sampling_size > data_frame.shape[0]:
        raise Exception(
            'The number of sample of sample is bigger than the shape of D')
    data_frame_sample = data_frame.sample(n=sampling_size)
    # Get the distance to their neirest neighbors in D : X
    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(data_frame_sample, k=2)
    data_frame_sample_distances_to_nearest_neighbours = dist[:, 1]
    # Randomly simulate n points with the same variation as in D : Q.
    max_data_frame = data_frame.max()
    min_data_frame = data_frame.min()
    uniformly_selected_values_0 = np.random.uniform(min_data_frame[0], max_data_frame[0], sampling_size)
    uniformly_selected_values_1 = np.random.uniform(min_data_frame[1], max_data_frame[1], sampling_size)
    uniformly_selected_observations = np.column_stack((uniformly_selected_values_0, uniformly_selected_values_1))
    if len(max_data_frame) >= 2:
        for i in range(2, len(max_data_frame)):
            uniformly_selected_values_i = np.random.uniform(min_data_frame[i], max_data_frame[i], sampling_size)
            to_stack = (uniformly_selected_observations, uniformly_selected_values_i)
            uniformly_selected_observations = np.column_stack(to_stack)
    uniformly_selected_observations_df = pd.DataFrame(uniformly_selected_observations)
    # Get the distance to their neirest neighbors in D : Y
    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(uniformly_selected_observations_df, k=1)
    uniformly_df_distances_to_nearest_neighbours = dist
    # return the hopkins score
    x = sum(data_frame_sample_distances_to_nearest_neighbours)
    y = sum(uniformly_df_distances_to_nearest_neighbours)
    if x + y == 0:
        raise Exception('The denominator of the hopkins statistics is null')
    return x / (x + y)[0]
# normalde bu fonksiyonumuz pyclustertend'ın hopkins kütüphanesinde var. Ancak oluşabilecek hatalara karşı
# doğrudan fonksiyonun kendisini kullanacağız. Bundan dolayı pyclustertend'ı install etmemize gerek yoktur!
                                                #!pip install pyclustertend --user
                                                # from pyclustertend import hopkins
```

```python
from sklearn.neighbors import BallTree
import numpy as np
import pandas as pd

def hopkins(data, sampling_size):
    """Assess the clusterability of a dataset. A score between 0 and 1.
    0.5 indicates no clusterability and a score close to 0 indicates high cluster tendency."""
    
    if type(data) == np.ndarray:
        data = pd.DataFrame(data)
    if sampling_size > data.shape[0]:
        raise Exception('Sampling size is larger than data shape.')
        
    sample = data.sample(n=sampling_size)
    tree = BallTree(data)
    _, true_dist = tree.query(sample, k=2)
    
    rand_data = pd.DataFrame({col: np.random.uniform(data[col].min(), data[col].max(), size=sampling_size) for col in data.columns})
    _, rand_dist = tree.query(rand_data)
    
    hopkins_stat = true_dist[:, 1].sum() / (true_dist[:, 1].sum() + rand_dist.sum())
    
    return hopkins_stat

```

# PCA's `fit_transform` Method

Using the code `principal_components = pca.fit_transform(scaled_X)`, we perform the two main steps of the PCA algorithm: `fit` and `transform`.

## 1. Fit Step:

In this step, PCA "fits" to the data and finds the principal components. The following processes occur during this step:

- **Centering:** The average of the features is calculated, and these averages are subtracted from each feature.
- **Calculation of the Covariance Matrix:** The covariance matrix is calculated based on the centered data.
- **Calculation of Eigenvalues and Eigenvectors:** The eigenvalues and eigenvectors of the covariance matrix are calculated. Eigenvectors indicate the directions of the principal components; eigenvalues indicate the variance of these components in the data.
- **Selection of Principal Components:** If not all principal components are desired (e.g., if the component count given to PCA is 2), eigenvectors with the largest eigenvalues are selected.

## 2. Transform Step:

In this step, PCA transforms the original data into a representation over the new principal components:

- **Resizing the Data:** Using the eigenvectors found in the `fit` step (directions of the principal components), the data is transformed into a new space. This transformation is obtained by multiplying the original data matrix with the eigenvectors.

The `fit_transform` method performs these two steps sequentially, resulting in the data being transformed into a representation over the principal components.

> **Note:** If you want to transform a new dataset using the same PCA model later on, you can use only the `transform` method on this new data. This is because the `fit` step is done only once, on the initial dataset.
